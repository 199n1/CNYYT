# 1. Introducción a los Números Complejos
    - Definición y notación
    - Partes real e imaginaria

Los números complejos son una extensión de los números reales y son fundamentales en matemáticas avanzadas e ingeniería. Proporcionan una solución a ecuaciones que no tienen soluciones reales, como la raíz cuadrada de números negativos.

## Definición

Un número complejo se define como un número de la forma $ a + bi $, donde:

- $ a $ y $ b $ son números reales,  
- $ i $ es la unidad imaginaria con la propiedad $ i² = -1 $.  

La parte real del número complejo es $ a $, y la parte imaginaria es $ b $
a+bi -> (a,b).  

## Unidad Imaginaria

La unidad imaginaria $ i $ se define por su propiedad $ i² = -1 $. Permite la definición de los números complejos y facilita el cálculo de raíces cuadradas de números negativos.  
Por ejemplo, la raíz cuadrada de -1 se representa como $ i $, y la raíz cuadrada de -4 es $ 2i $.  

## Plano Complejo

El plano complejo es un plano bidimensional donde se grafican los números complejos.  
El eje horizontal representa la parte real, y el eje vertical representa la parte imaginaria.  

## Aplicaciones

Los números complejos tienen aplicaciones en diversos campos, incluyendo ingeniería, física, ciencias de la computación y más.  
Son especialmente útiles para representar oscilaciones, ondas y otros fenómenos que involucran magnitudes bidimensionales.



# 2) Operaciones básicas con números complejos  

- **Suma, resta, multiplicación y división**  
- **Conjugado y módulo**  

## Operaciones algebraicas
Para $a+bi$ y $c+di$:  

- **Addition**: $+: \mathbb{C} \times \mathbb{C}   \rightarrow \mathbb{C}:  (a_1+b_1i) + (a_2+b_2i) = (a_1+a_2) + (b_1+b_2)i$
- **Subtraction**: $-: \mathbb{C} \times \mathbb{C}   \rightarrow \mathbb{C}: (a + bi) - (c + di) = (a - c) + (b - d)i $
- **Multiplication**: $+: \mathbb{C} \times \mathbb{C}   \rightarrow \mathbb{C}: (a + bi)(c + di) = (ac - bd) + (ad + bc)i $
- **Division**: $/: \mathbb{C} \times \mathbb{C}   \rightarrow \mathbb{C}: \frac{a + bi}{c + di} = \frac{(ac + bd) + (bc - ad)i}{c² + d²} $

## Complex Conjugate

El conjugado de $a+bi$ es $a-bi$: se cambia el signo de la parte imaginaria.  
Es útil para divisiones y para calcular el módulo.  The complex conjugate of a complex number \( a + bi \) is \( a - bi \). It is obtained by changing the sign of the imaginary part. This concept is particularly useful in division and in finding the magnitude of complex numbers.

$$ \text{ }^{-} : \mathbb{C} \rightarrow \mathbb{C} : \bar c = \overline {a + bi} = a -bi $$

## Magnitude (Modulus)

The magnitude (or modulus) of a complex number \( a + bi \) is given by \( \sqrt{a² + b²} \). It represents the distance of the complex number from the origin in the complex plane.

$$| \text{ }| : \mathbb{C} \rightarrow \mathbb{R} : |c| = |a + bi| = + \sqrt{a^2+b^2}$$

# 4. Polar Representation of Complex Numbers

In addition to the rectangular form (a + bi), complex numbers can also be represented in polar form. The polar form of a complex number expresses it in terms of a magnitude (r) and an angle (θ), and is written as `r(cos θ + i sin θ)` or `r∠θ`. This form is particularly useful in quantum computing, signal processing, and other fields where phase and amplitude are important.

## Key Concepts
- **Magnitude (r)**: The distance of the complex number from the origin in the complex plane, calculated as `sqrt(a^2 + b^2)`.
- **Angle (θ)**: The counterclockwise angle measured from the positive x-axis to the line segment that joins the point to the origin, often given in radians.
- **Euler's Formula**: `e^(iθ) = cos θ + i sin θ`, which allows the polar form to be written as `r * e^(iθ)`.

## Converting between Rectangular and Polar Forms
- To convert from rectangular to polar form, use `r = sqrt(a^2 + b^2)` and `θ = atan2(b, a)`.
- To convert from polar to rectangular form, use `a = r * cos(θ)` and `b = r * sin(θ)`.

## Python Implementation
Use Python's `cmath` module for working with complex numbers in polar form.

## An interactive tool for polar and cartesian conversion

````python
import cmath
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def convert_and_plot(real=0, imaginary=0, magnitude=0, angle=0, convert_to='polar'):
    if convert_to == 'polar':
        z = complex(real, imaginary)
        r, theta = cmath.polar(z)
        display_text = f"Cartesian: ({real}, {imaginary}) -> Polar: (r={r:.2f}, θ={np.degrees(theta):.2f}°)"
    else:
        z = cmath.rect(magnitude, np.radians(angle))
        display_text = f"Polar: (r={magnitude}, θ={angle}°) -> Cartesian: ({z.real:.2f}, {z.imag:.2f})"

    fig, ax = plt.subplots()
    ax.plot([0, z.real], [0, z.imag], 'ro-')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_xlabel('Real')
    ax.set_ylabel('Imaginary')
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    ax.grid(True)
    plt.title(display_text)
    plt.show()

style = {'description_width': 'initial'}
layout = widgets.Layout(width='400px')

# Creating interactive widgets
real_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=0, description='Real Part:', style=style, layout=layout)
imaginary_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=0, description='Imaginary Part:', style=style, layout=layout)
magnitude_slider = widgets.FloatSlider(min=0, max=5, step=0.1, value=1, description='Magnitude:', style=style, layout=layout)
angle_slider = widgets.FloatSlider(min=-180, max=180, step=1, value=0, description='Angle (degrees):', style=style, layout=layout)
convert_dropdown = widgets.Dropdown(options=['polar', 'cartesian'], value='polar', description='Convert to:', style=style, layout=layout)

# Interactive function

ui = widgets.VBox([widgets.HBox([real_slider, imaginary_slider]),
                   widgets.HBox([magnitude_slider, angle_slider]),
                   convert_dropdown])
out = widgets.interactive_output(convert_and_plot, {'real': real_slider, 'imaginary': imaginary_slider,
                                                    'magnitude': magnitude_slider, 'angle': angle_slider,
                                                    'convert_to': convert_dropdown})

display(ui, out)

````

## Euler's Notation for Complex Numbers

Euler's formula is a fundamental equation in complex analysis and provides a deep connection between trigonometry and complex exponential functions. It states that for any real number `x`, the following identity holds:

$$ e^{ix} = \cos(x) + i\sin(x) $$

This equation is known as Euler's formula. In this formula, `e` is the base of the natural logarithm, `i` is the imaginary unit, and `cos` and `sin` are the trigonometric functions cosine and sine respectively. The formula is often used in the polar representation of complex numbers.

### Polar Representation Using Euler's Formula

A complex number can be represented in polar coordinates as `r(cos θ + i sin θ)`. Using Euler's formula, this can be written in the more compact form:

$$ z = r \cdot e^{iθ} $$

where `r` is the magnitude (modulus) of the complex number, and `θ` is the angle formed with the positive real axis, often referred to as the argument of the complex number.

### Applications

Euler's formula is widely used in various fields such as electrical engineering, quantum mechanics, and signal processing. It simplifies the analysis of waveforms, oscillations, and rotating systems.

## Understanding the Mandelbrot Set

The Mandelbrot Set is a fascinating and complex structure in the field of fractal geometry, discovered by mathematician Benoit Mandelbrot in 1980. It's a set of complex numbers that forms an intricate and infinitely detailed boundary when plotted in the complex plane.

### What is the Mandelbrot Set?
The Mandelbrot Set is defined as the set of complex numbers `c` for which the function `f(z) = z^2 + c` does not diverge when iterated from `z = 0`. In simpler terms, if repeatedly squaring the number and adding `c` never makes the value go to infinity, that number is part of the Mandelbrot Set.

### Key Concepts:
- **Iteration**: Starting with `z = 0`, we repeatedly apply the function `f(z) = z^2 + c`. The behavior of this sequence determines if `c` is in the Mandelbrot Set.
- **Boundedness**: If the sequence remains bounded (doesn't go to infinity), then `c` is in the Mandelbrot Set.
- **Escape Time Algorithm**: To visualize the set, we use an algorithm that assigns a color based on how quickly the values reach a certain threshold when iterated. Points inside the set typically remain black.

### Properties of the Mandelbrot Set:
- **Self-Similarity**: The set is fractal, meaning it contains smaller copies of itself.
- **Complex Boundary**: The boundary of the Mandelbrot Set is infinitely complex. Zooming into the boundary reveals more detail.
- **Area**: While the exact area of the Mandelbrot Set is unknown, it is finite.
- **Connectedness**: Despite its complexity, the set is mathematically proven to be a connected set.

### Plotting the Mandelbrot Set:
To visualize the set, we iterate over a grid of complex numbers and use the escape time algorithm to color each point. Points that remain bounded (do not escape) after a set number of iterations are considered part of the Mandelbrot Set and are usually colored black.

The beauty of the Mandelbrot Set lies not just in its visual complexity, but also in the way it bridges mathematics, art, and computer science. It serves as a powerful example of how simple rules can lead to incredibly intricate structures.


# 5. Aplicación: Fractales y Números Complejos  

## 5.1 Introducción a los fractales  
Un **fractal** es una figura geométrica que presenta una estructura compleja e infinita en detalle.  
Los fractales se caracterizan por:  
- **Autosimilitud**: contienen copias más pequeñas de sí mismos.  
- **Complejidad infinita**: al hacer zoom, aparecen más detalles sin límite.  

Uno de los fractales más famosos es el **Conjunto de Mandelbrot**.  

---

## 5.2 Comprendiendo el Conjunto de Mandelbrot  

El **Conjunto de Mandelbrot** fue descubierto en 1980 por el matemático **Benoît Mandelbrot**.  
Es un conjunto de números complejos que, al graficarse en el plano complejo, forma una frontera increíblemente intrincada y detallada.  

### Definición  
El Conjunto de Mandelbrot está formado por los números complejos $c$ para los cuales la función:  

$$ f(z) = z^2 + c $$  

no diverge (no se va al infinito) cuando se itera a partir de $z_0 = 0$.  

En términos simples:  
- Si al **elevar al cuadrado y sumar $c$ repetidamente** el resultado nunca se escapa al infinito,  
- entonces $c$ pertenece al Conjunto de Mandelbrot.  

---

## 5.3 Conceptos clave  

- **Iteración**: comenzamos con $z_0 = 0$ y aplicamos $f(z) = z^2 + c$ varias veces.  
- **Acotación**: si la secuencia permanece acotada (no explota a infinito), $c$ está en el conjunto.  
- **Algoritmo del Tiempo de Escape**: para graficar el conjunto, se colorea cada punto según qué tan rápido se escapa hacia el infinito.  
  - Los puntos que no escapan (pertenecen al conjunto) se dibujan en **negro**.  

---

## 5.4 Propiedades del Conjunto de Mandelbrot  

- **Autosimilitud**: contiene copias más pequeñas de sí mismo.  
- **Frontera compleja**: la frontera del conjunto es infinitamente intrincada.  
- **Área finita**: aunque su borde es infinito en detalle, el área total es finita.  
- **Conectividad**: está matemáticamente demostrado que el conjunto es **conexo** (toda su estructura está unida).  

---

## 5.5 Visualización  

Para graficar el Conjunto de Mandelbrot se realiza:  
1. Crear una **rejilla de números complejos**.  
2. Iterar la función $f(z) = z^2 + c$.  
3. Usar el **algoritmo del tiempo de escape** para colorear cada punto.  

🔲 Los puntos dentro del conjunto → **negro**.  
🌈 Los puntos fuera del conjunto → colores según la rapidez con que divergen.  

---

✨ La belleza del Conjunto de Mandelbrot radica en que **reglas simples** ($z^2 + c$) generan **estructuras infinitamente complejas**, conectando matemáticas, arte y computación.


````python
# Example: Generating a Mandelbrot set

import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Create a grid of complex numbers
pixels = 800
max_iter = 30
x = np.linspace(-2, 1, pixels)
y = np.linspace(-1.5, 1.5, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

# Compute Mandelbrot set
mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

# Plot
plt.imshow(mandelbrot_set, extent=[-2, 1, -1.5, 1.5])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
````

````python
import matplotlib.pyplot as plt

def plot_vector(ax, z, label=None):
    """Dibuja un vector desde el origen hasta el punto (Re(z), Im(z))."""
    ax.plot([0, z.real], [0, z.imag], marker='o')
    if label:
        ax.annotate(label, xy=(z.real, z.imag), xytext=(5,5), textcoords='offset points')

def preparar_ejes(ax, limite=6):
    """Configura los ejes del plano complejo."""
    ax.set_xlim(-limite, limite)
    ax.set_ylim(-limite, limite)
    ax.set_xlabel("Parte real")
    ax.set_ylabel("Parte imaginaria")
    ax.axhline(0, linewidth=0.5)
    ax.axvline(0, linewidth=0.5)
    ax.grid(True)

#  Grafica (3+4j) y su conjugado
z = 3 + 4j
fig, ax = plt.subplots(figsize=(5,5))
plot_vector(ax, z, "z = 3+4j")
plot_vector(ax, z.conjugate(), "conj(z) = 3-4j")
preparar_ejes(ax, limite=6)
ax.set_title("Número complejo y su conjugado")
plt.show()

# Graficar suma y multiplicación
zA = 3 + 4j
zB = 1 - 2j
suma_AB = zA + zB
producto_AB = zA * zB

fig, ax = plt.subplots(figsize=(6,6))
plot_vector(ax, zA, "zA")
plot_vector(ax, zB, "zB")
plot_vector(ax, suma_AB, "zA + zB")
plot_vector(ax, producto_AB, "zA * zB")
preparar_ejes(ax, limite=8)
ax.set_title("Suma y multiplicación en el plano complejo")
plt.show()
````

# Ejercicio 3: Conjunto de Mandelbrot

````python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot_set_vectorized(xmin, xmax, ymin, ymax, width, height, max_iter):
    """Genera la matriz de escape del conjunto de Mandelbrot."""
    xs = np.linspace(xmin, xmax, width)
    ys = np.linspace(ymin, ymax, height)
    C = xs[np.newaxis, :] + 1j * ys[:, np.newaxis]
    Z = np.zeros_like(C, dtype=complex)
    output = np.zeros(C.shape, dtype=int)

    for n in range(max_iter):
        mask = np.abs(Z) <= 2
        Z[mask] = Z[mask]**2 + C[mask]
        escaped = (mask) & (np.abs(Z) > 2)
        output[escaped] = n + 1

    return output

# Parámetros para zoom y centro
center = -0.75 + 0.1j
zoom = 0.6
span_x = 3.0 * zoom
span_y = 3.0 * zoom

xmin = center.real - span_x/2
xmax = center.real + span_x/2
ymin = center.imag - span_y/2
ymax = center.imag + span_y/2

mandel = mandelbrot_set_vectorized(xmin, xmax, ymin, ymax, width=600, height=400, max_iter=100)

plt.figure(figsize=(8,5))
plt.imshow(mandel, extent=[xmin, xmax, ymin, ymax], origin='lower')
plt.colorbar(label='Iteraciones antes de escapar')
plt.title(f"Mandelbrot (centro={center}, zoom={zoom})")
plt.show()
````

# Ejercicio 4: Conjunto de Julia

````python
import numpy as np
import matplotlib.pyplot as plt

def julia_set(c, xmin, xmax, ymin, ymax, width, height, max_iter):
    """Genera el conjunto de Julia para la constante c."""
    xs = np.linspace(xmin, xmax, width)
    ys = np.linspace(ymin, ymax, height)
    Z = xs[np.newaxis, :] + 1j * ys[:, np.newaxis]
    output = np.zeros(Z.shape, dtype=int)

    for n in range(max_iter):
        mask = np.abs(Z) <= 2
        Z[mask] = Z[mask]**2 + c
        escaped = (mask) & (np.abs(Z) > 2)
        output[escaped] = n + 1

    return output

# Constante de Julia (se puede probar otras)
c = -0.4 + 0.6j

julia = julia_set(c, xmin=-1.5, xmax=1.5, ymin=-1.5, ymax=1.5,
                  width=600, height=600, max_iter=150)

plt.figure(figsize=(6,6))
plt.imshow(julia, extent=[-1.5, 1.5, -1.5, 1.5], origin='lower')
plt.colorbar(label='Iteraciones antes de escapar')
plt.title(f"Conjunto de Julia (c = {c})")
plt.show()
````

# Ejercicio 5: Propiedades de números complejos

````python

# Propiedad A: |z*w| = |z| * |w|
z = 1 + 2j
w = -0.5 + 3j

lhs_mod = abs(z * w)      # lado izquierdo
rhs_mod = abs(z) * abs(w) # lado derecho

print("=== Ejercicio 5 ===")
print(f"z = {z}, w = {w}")
print(f"|z*w| = {lhs_mod:.12f}")
print(f"|z| * |w| = {rhs_mod:.12f}")

# Propiedad B: conj(a+b) = conj(a) + conj(b)
a = 2 + 3j
b = -1 + 4j

lhs_conj = (a + b).conjugate()
rhs_conj = a.conjugate() + b.conjugate()

print()
print(f"a = {a}, b = {b}")
print(f"conj(a+b) = {lhs_conj}")
print(f"conj(a) + conj(b) = {rhs_conj}")
````
